<h1>Exploratory Data Analysis On Reserve Traffic</h1>

<ul>
    <li><h3>Imports</h3></li>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style('whitegrid')


%matplotlib inline
```

<ul>
    <li><h3>Load data</h3></li>
</ul>

```python
dataset = pd.read_csv('LekagulSensorData.csv', parse_dates=["Timestamp"],
                      date_parser=lambda x: pd.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

<ul>
    <li><h3>Traffic volume analysis by car type</h3></li>
</ul>

```python
dataset.reset_index()
counts = dataset.groupby("car-type").count().sort_values(by='Timestamp', ascending=False)

fig = sns.barplot(data=counts, x='car-id', y=counts.index)
fig.set(xlabel='Traffic Volume', ylabel='Car Type')

plt.show()
```

<ul>
    <li><h3>Traffic volume analysis by gate name</h3></li>
</ul>

```python
dataset.reset_index()
counts = dataset.groupby("gate-name").count().sort_values(by='Timestamp', ascending=False)

fig = sns.barplot(data=counts, x='car-id', y=counts.index)
fig.set(xlabel='Traffic Volume', ylabel='Gate Name')
fig.figure.set_size_inches(18,30)

plt.show()
```
